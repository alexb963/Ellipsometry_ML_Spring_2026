# The goal of this workbook is to read in ellipsometry data and the corresponding JSON files in order to get the "input data" (SE data) and "output data" (Layers, Thicknesses, Compositions...) to train a neural network

```python
def read_SE_data(path):

    """
    
    This function will need a folder path as an input and will produce a set of SE data along with the information needed to 
     generate the SE data. The folder is expected to contain csv files containing the SE data and JSON files containing the 
     infromation needed to simulate the SE data
     
    """

    # Step 1: get CSV and JSON file names 

    csv_file = glob.glob(path + "/*.csv")
    JSON_file = glob.glob(path + "/*.JSON")

    # Step 1.5: Read the files 

    SE_data = [] # list to store SE data
    Meta_data = [] # list to store meta data

    L1_Thickness = []
    L2_Thickness = []
    L1_Composition = []

    for i in range(len(csv_file)): 

        df = pd.read_csv(csv_file[i]) # Load in csv file
        # Step 2: Drop 1st file of CSV column
        df.drop(df.columns[0], axis=1, inplace=True) # remove unwanted column 
        df.drop(df.columns[0], axis=1, inplace=True) # remove Wavele
        df.to_numpy()
        SE_data.append(df) # save SE data 

        df = pd.read_json(JSON_file[i]) # Load in JSON file

        # Read Composition 
        Composition = float(df['Layer Info']['Layer1']['Tellurium_Content(%)'])

        # Read Thicknesses 

        L1_T = float(df['Layer Info']['Layer1']['thickness_nm'])
        L2_T = float(df['Layer Info']['Layer2']['thickness_nm'])

        L1_Thickness.append(L1_T) 
        L2_Thickness.append(L2_T)

        
        # convert composition to binary categorical data


        if Composition == 86:

            arr = np.array([1, 0, 0,   # convert 86 to binary category
                            0, 0, 0, 
                            0, 0, 0])
    
            L1_Composition.append(arr)

    
        if Composition == 75:

            arr = np.array([0, 1, 0,   # convert 75 to binary category
                            0, 0, 0, 
                            0, 0, 0])
    
            L1_Composition.append(arr)


        if Composition == 69:

            arr = np.array([0, 0, 1,   # convert 69 to binary category
                            0, 0, 0, 
                            0, 0, 0])
    
            L1_Composition.append(arr)

            
        if Composition == 64:

            arr = np.array([0, 0, 0,   # convert 64 to binary category
                            1, 0, 0, 
                            0, 0, 0])
    
            L1_Composition.append(arr)
        
        
        if Composition == 54:

            arr = np.array([0, 0, 0,   # convert 54 to binary category
                            0, 1, 0, 
                            0, 0, 0])
    
            L1_Composition.append(arr)

        if Composition == 38:

            arr = np.array([0, 0, 0,   # convert 54 to binary category
                            0, 0, 1, 
                            0, 0, 0])
    
            L1_Composition.append(arr)

        if Composition == 25:

            arr = np.array([0, 0, 0,   # convert 54 to binary category
                            0, 0, 0, 
                            1, 0, 0])
    
            L1_Composition.append(arr)

        if Composition == 15:

            arr = np.array([0, 0, 0,   # convert 54 to binary category
                            0, 0, 0, 
                            0, 1, 0])
    
            L1_Composition.append(arr)

        if Composition == 0:

            arr = np.array([0, 0, 0,   # convert 54 to binary category
                            0, 0, 0, 
                            0, 0, 1])
    
            L1_Composition.append(arr)






        
        #Meta_data.append(df) # Optionally save JSON file

        





    # Put SE data into an array along with thicknesses and composition
    SE_data = np.array(SE_data)
    L1_Thickness = np.array(L1_Thickness)
    L2_Thickness = np.array(L2_Thickness)
    L1_Composition = np.array(L1_Composition)

    # Put Thickness and Composition into array



    return(SE_data, L1_Thickness, L2_Thickness, L1_Composition)
```

```python
def unison_shuffled_copies(*arrays):
    # Ensure all input arrays have the same length
    lengths = [len(arr) for arr in arrays]
    assert all(length == lengths[0] for length in lengths), "All arrays must have the same length"
    
    # Generate a random permutation of indices
    p = np.random.permutation(lengths[0])
    
    # Shuffle each array according to the permutation
    return tuple(arr[p] for arr in arrays)
```

```python
import tensorflow as tf
from tensorflow import keras
import tensorflow.keras.layers as tfl
```

```python
x_train = SE_data_train
y_train = L1_Composition_train

x_val = SE_data_val
y_val = L1_Composition_val
```

```python
# Now to build the model 

# Define the Input layer of the ANN.
input_shape = (697, 3) # 697 data points with the 3 columns being  ( N, C, S)
input_NCS = tf.keras.Input(shape=input_shape)

x = tfl.Flatten()(input_NCS) 

x = tfl.Dense(units=128, activation='leaky_relu')(x)
x = tfl.Dense(units=64, activation='leaky_relu')(x)
x = tfl.Dense(units=32, activation='leaky_relu')(x)
x = tfl.Dense(units=16, activation='leaky_relu')(x)

# Now define the output layer. The softmax activation function is chosen because it is bound between [0,1]. Perfect for binary categorical problems. 
x = tfl.Dense(units=9, activation='softmax')(x)


# Create the model
model = tf.keras.Model(inputs=input_NCS, outputs= x)
```

```python
model.summary()
```

```python

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss =  tf.keras.losses.CategoricalCrossentropy()
)

# Train the model.
history = model.fit(
    x=x_train, # Input data
    y=y_train, # Output data 
    batch_size=32, # Batch size (adjust as needed)
    epochs=50, # Number of epochs (adjust as needed)
    validation_data=(x_val, y_val), # Validation data
    verbose=1, # Verbosity mode
    #callbacks=[early_stopping] # Ends training when performance on validation data stops improving
)
```

```python
x_test = SE_data_test
y_test = L1_Composition_test

model.evaluate(x_test, y_test)
```
